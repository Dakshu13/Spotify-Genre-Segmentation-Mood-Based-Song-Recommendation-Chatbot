```python
from sklearn.cluster import KMeans
```

```python
knn_model=KMeans(n_clusters=6)
```

```python
knn_model.fit(a[["danceability","key"]])
```

```python
knn_model.cluster_centers_
```

```python
a["clustergroup"]=knn_model.labels_
a
```

```python
genre_labels = {
    0: "Pop",
    1: "Rap",
    2: "Rock",
    3: "Latin",
    4: "R&B",
    5: "EDM"
}
```

```python
a["genre_cluster"] = a["clustergroup"].map(genre_labels)
a
```

```python
sns.scatterplot(x="danceability",y="key",data=a,hue="genre_cluster")
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(a[["danceability","key"]], a["clustergroup"])
print("Score:", score)
```

```python
# Count the number of songs per genre cluster
genre_counts = a["genre_cluster"].value_counts()

# Simple barh plot
clrs=['red','blue','green','yellow','cyan','purpl']
plt.barh(genre_counts.index, genre_counts.values,color=clrs)

plt.xlabel("Number of Songs")
plt.ylabel("Genre")
plt.title("Songs per Genre Cluster")
plt.show()
```

```python
# Mood-based chatbot to suggest first 10 songs from genre clusters

def chatbot_cluster_response(cluster_id):
    if cluster_id == 0:
        print("🎉 You’re in the party mood! Try these songs:")
        for song in a[a['clustergroup'] == 0]['track_name'].head(10):
            print("-", song)
    elif cluster_id == 1:
        print("💤 Feeling chill? These tracks are perfect:")
        for song in a[a['clustergroup'] == 1]['track_name'].head(10):
            print("-", song)
    elif cluster_id == 2:
        print("🔥 Time to get energized! Check out these hits:")
        for song in a[a['clustergroup'] == 2]['track_name'].head(10):
            print("-", song)
    elif cluster_id == 3:
        print("💔 In your feels? These emotional tracks might help:")
        for song in a[a['clustergroup'] == 3]['track_name'].head(10):
            print("-", song)
    elif cluster_id == 4:
        print("🌞 Upbeat vibes! Here’s your playlist:")
        for song in a[a['clustergroup'] == 4]['track_name'].head(10):
            print("-", song)
    elif cluster_id == 5:
        print("🌙 Late night moods, coming right up:")
        for song in a[a['clustergroup'] == 5]['track_name'].head(10):
            print("-", song)
    else:
        print("😕 Hmm... that cluster doesn't exist. Try a number between 0 and 5.")

# User input to get a song list from the selected mood cluster
try:
    a1 = input("Hello vibey!!, what's your name?? ")
    print(f"Nice to meet you, {a1}!")
    a2 = input("Hi there! 🎧 How's your day going? ")
    print("That's great to hear! 🎶")

    print("What kind of music are you interested in? (pop, rap, rock, latin, r&b, edm):")
    mood = input("Enter your mood: ").lower()

    mood_map = {
        'pop': 0,
        'rap': 1,
        'rock': 2,
        'latin': 3,
        'r&b': 4,
        'edm': 5
    }

    if mood in mood_map:
        user_cluster = mood_map[mood]
        chatbot_cluster_response(user_cluster)
    else:
        print("😕 Mood not recognized. Please choose from: pop, rap, rock, latin, r&b, edm.")

except:
    print("Please enter a valid input.")
```
